```python
# diagnostic_individual.py
import pandas as pd
import numpy as np

# Load your individual data (from data_merge.py output)
df = pd.read_csv('../data/processed/brfss_individual_cleaned.csv')
print(f"Individual records: {len(df):,}")
print(f"Columns: {len(df.columns)}")
print("\nFirst 5 rows:")
print(df.head())
print("\nDisease columns (binary 0/1):")
disease_cols = [col for col in df.columns if any(disease in col for disease in ['Diabetes', 'HeartDisease', 'Stroke', 'Asthma', 'COPD'])]
for col in disease_cols:
    if col in df.columns:
        print(f"{col}: {df[col].sum():,} cases ({df[col].mean()*100:.1f}%)")

print("\nUnique counties (FIPS):", df['FIPS'].nunique())
print("Unique states:", df['State'].nunique())
```
